```python
#importação das bibliotecas
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import LabelEncoder
from datetime import datetime
import matplotlib.pyplot as plt
from collections import defaultdict
%matplotlib inline
```

Para o dataset trade sera necessário transformar o campo data em tipo date, em seguida reorganizamos e reordenamos o dataset deixando o campo date como index.

```python
#leitura do data set trade
trade = pd.read_csv('trade.csv')

# trade['dataFormatada'] = str(trade['Date'][0:11])

def calculadata(coluna):
    lista = []
    for x in coluna:    
        numeros = x[0:10].split('-')   
        x = int(numeros[0])+int(numeros[1])*30+int(numeros[2])*365
        lista.append(x - 624380)
    return lista

trade['Data_enc'] = calculadata(trade['Date'])
trade.sort_values('Data_enc', inplace = True)
# trade.hist()
trade.head()
trade.tail()
min(trade['Close'])
```

```python
from sklearn import datasets, linear_model

X = trade['Data_enc'].values.reshape(-1,1)
y = trade['Close'].values.reshape(-1,1)

print(y)


regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(X, y)

predict = regr.predict(X)
```

```python
plt.plot(X, predict, color='r')
plt.plot(X, y)
```

```python
from sklearn.svm import SVR

amostras = 100

X = X[:amostras]
y = y[:amostras]


svr = SVR(kernel='poly')
svr.fit(X, y)

predictSVR = svr.predict(X)
plt.plot(X, y)
plt.plot(X, predictSVR, color='r')
```

Para o dataset dinos foi feito um Label Encoder para transformar variaveis classificatorias em numericas

```python
#leitura do dataset dinos
dinos = pd.read_csv('dinos.csv')
#dinos = dinos.iloc[:,1:]

#tratamento do dataset dinos
label = LabelEncoder()


dinos['workclass'] = label.fit_transform(dinos['workclass'])
dinos['education_level'] = label.fit_transform(dinos['education_level'])
dinos['marital-status'] = label.fit_transform(dinos['marital-status'])
dinos['occupation'] = label.fit_transform(dinos['occupation'])
dinos['relationship'] = label.fit_transform(dinos['relationship'])
dinos['race'] = label.fit_transform(dinos['race'])
dinos['sex'] = label.fit_transform(dinos['sex'])
dinos['native-country'] = label.fit_transform(dinos['native-country'])

dinos_enconder = dinos
dinos.head()
#dinos.to_csv('dinos_encoder.csv')
```
